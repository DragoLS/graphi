```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

terra = pd.read_csv("Terraria DPS_TV1.4.4.9_V1 - Sheet1.csv")

y = [5,4,3,2,1]
x = terra["DPS (SINGLE TARGET)"].head(5)
z = terra["DPS (MULTI TARGET)"].head(5)
terr=terra.replace({"DPS (SINGLE TARGET + PROYECTILE ONLY)":np.NaN},0)
c = terr["DPS (MULTI TARGET + PROYECTILE ONLY)"].head(5)
plt.plot(y,x,label = "     ")
plt.plot(y,z,label = "     ")
plt.plot(y,c,label = "     ")
plt.legend()
plt.show()
```
